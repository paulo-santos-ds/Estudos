```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
# Conectar ao banco de dados (no exemplo, na memória)
conn = sqlite3.connect(':memory:')

# Criar uma tabela
conn.execute('''CREATE TABLE IF NOT EXISTS exemplo (
                id INTEGER PRIMARY KEY,
                nome TEXT NOT NULL,
                idade INTEGER NOT NULL)''')

# Inserir dados na tabela
conn.execute("INSERT INTO exemplo (nome, idade) VALUES ('João', 30)")
conn.execute("INSERT INTO exemplo (nome, idade) VALUES ('Maria', 25)")

# Consultar dados
cursor = conn.execute("SELECT * FROM exemplo")
for linha in cursor:
    print(linha)

# Fechar a conexão
conn.close()

```
